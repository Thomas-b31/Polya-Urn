```python
import numpy as np
import matplotlib.pyplot as plt


def une_trajectoire(n, r, nb_pas):
    if nb_pas == 0:
        return [n / (n + r)]
    elif np.random.randint(low=1, high=n + r + 1) > n:  # génère un nombre aléatoire et décide d'incrémenter n ou r en conséquence
        return [n / (n + r)] + une_trajectoire(n, r + 1, nb_pas - 1)
    else:
        return [n / (n + r)] + une_trajectoire(n + 1, r, nb_pas - 1)


for i in range(5):
    une_trajectoire(1, 1, 10)  # génère et affiche cinq trajectoires avec une longueur de 10 pas

print(une_trajectoire(1, 1, 10))  # affiche une trajectoire avec une longueur de 10 pas

plt.hist(une_trajectoire(1, 1, 10))  # crée un histogramme basé sur une trajectoire avec une longueur de 10 pas
```

```python
import numpy as np
import matplotlib.pyplot as plt


def une_trajectoire(n, r, nb_pas):
    # initialise la trajectoire avec le premier ratio n / (n + r)
    trajectoire = [n / (n + r)]

    # génère la trajectoire en fonction des déplacements aléatoires de n et r
    for _ in range(nb_pas):
        # compare un nombre aléatoire avec la somme de n et r pour décider d'incrémenter n ou r
        if np.random.randint(1, n + r + 1) > n:
            r += 1
        else:
            n += 1

        # ajoute le ratio actuel à la trajectoire
        trajectoire.append(n / (n + r))

    return trajectoire


# génère une trajectoire avec une longueur de 10 pas
une_trajectoire(1, 1, 10)

# génère et affiche 500 trajectoires avec une longueur de 75 pas
for i in range(500):
    plt.plot(une_trajectoire(1, 1, 75))
```

```python
pas = 75  # nombre de pas
real = 1000  # nombre d'itérations

# initialise un tableau avec des 0 pour stocker les états finaux des trajectoires
etats_finaux = np.zeros(real)

for i in range(real):
    # trajectoire avec 10 boules noires et 10 boules rouges
    traj = une_trajectoire(2, 2, pas)

    # stocke la probabilité à la dernière étape de la trajectoire
    etats_finaux[i] = traj[pas]

# affiche l'histogramme des états finaux
plt.hist(etats_finaux)
```

```python
import scipy.stats as st
st.probplot(etats_finaux, dist="norm", plot=plt)  # créé un graphique de probabilité des états finaux par rapport à une distribution normale
st.norm.fit(etats_finaux)  # donne les paramètres de notre loi et ajuste à une distribution normale
```

```python
from scipy import stats

a, b, loc, scale = st.beta.fit(etats_finaux)  # ajuste une distribution bêta aux données des états finaux
stats.probplot(etats_finaux, dist="beta", sparams=(a, b, loc, scale), plot=plt)  # créé un graphique de probabilité des états finaux par rapport à une distribution bêta
```

Si on ajoute 3 boules supplémentaires

```python
import numpy as np
import matplotlib.pyplot as plt


def une_trajectoire_3_boules(n, r, nb_pas):  # même fonction que précédemment mais en ajoutant 3 boules
    trajectoire = [n / (n + r)]

    for _ in range(nb_pas):
        if np.random.randint(1, n + r + 1) > n:
            r += 3
        else:
            n += 3

        trajectoire.append(n / (n + r))

    return trajectoire


# génère une trajectoire avec une longueur de 10 pas
une_trajectoire(1, 1, 10)

# génère et affiche 100 trajectoires avec une longueur de 20 pas
for i in range(100):
    plt.plot(une_trajectoire_3_boules(1, 1, 20))
```

```python
# Est ce que la loi Beta décrit correctement cette situation ?

pas = 25  # nombre de pas
real = 100  # nombre d'itérations

# initialise un tableau avec des 0 pour stocker les états finaux des trajectoires
etats_finaux_3_boules = np.zeros(real)


for i in range(real):
    traj = une_trajectoire_3_boules(1, 1, pas)
    etats_finaux_3_boules[i] = traj[pas]

# État final = dernière probabilité
plt.subplot(211)
plt.hist(etats_finaux_3_boules)

a, b, loc, scale = stats.beta.fit(etats_finaux_3_boules)
plt.subplots_adjust(hspace=0.4)  # ajuste l'espacement vertical entre les sous-graphiques
plt.subplot(212)
stats.probplot(etats_finaux_3_boules, dist="beta", sparams=(a, b, loc, scale), plot=plt)  # créé un graphique de probabilité des états finaux avec 3 boules par rapport à une distribution bêta
```
